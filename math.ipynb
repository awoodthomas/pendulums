```python
from sympy import symbols
import sympy
```

```python
n = 2
m = sympy.IndexedBase('m')
r = sympy.IndexedBase('r')
theta = sympy.IndexedBase('theta')
omega = sympy.IndexedBase('omega')
k, x = symbols('k x', integer=True)
g = symbols('g')

mass_matrix = sympy.Matrix([[sympy.summation(m[k], (k,max(i,j),n)) * r[i] * r[j] * sympy.cos(theta[i] - theta[j]) for j in range(n)] for i in range(n)])
mass_matrix
```

```python
# b = -np.array([np.sum([np.sum(m[max(i, j):]) * r[i] * r[j] * omega[j]**2 * math.sin(theta[i]-theta[j])
#                          for j in range(n)]) + g * np.sum(m[i:]) * r[i] * math.sin(theta[i]) for i in range(n)])
b = -sympy.Matrix([
    sympy.summation(sympy.summation(m[k], (k,sympy.Max(i,x),n-1)) * r[i] * r[x] * omega[x]**2 * sympy.sin(theta[i]-theta[x]),(x,0,n)) 
    + g * sympy.summation(m[k], (k,i,n-1)) * r[i] * sympy.sin(theta[i]) 
    for i in range(n)]).doit()
b
```

```python
from sympy.printing.numpy import NumPyPrinter
print("M: ", NumPyPrinter().doprint(mass_matrix))
print("b: ", NumPyPrinter().doprint(b))
```

```python

transform = mass_matrix.LUsolve(b)
transform = sympy.simplify(transform)
transform
```
